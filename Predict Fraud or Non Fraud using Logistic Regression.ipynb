```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../Data/creditcard.csv")
```

```python
df=df.drop(columns="Time")
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df["Amount"]=sc.fit_transform(df[["Amount"]])
```

```python
x=df.drop(columns=["Class"])
y=df["Class"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
acc=accuracy_score(ytest,ypred)
print("Accuracy is :",acc)
cm=confusion_matrix(ytest,ypred)
sns.heatmap(cm,annot=True)
print(classification_report(ytest,ypred))
```

### Under Sampling


```python
nonfraud=df[df.Class==0].index
fraud=np.array(df[df.Class==1].index)

random_index=np.array(np.random.choice(nonfraud,len(fraud),replace=False))
undersample_index=np.concatenate([fraud,random_index])
undersample=df.iloc[undersample_index,:]
```

```python
x=undersample.drop(columns=["Class"])
y=undersample["Class"]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
acc=accuracy_score(ytest,ypred)
print("Accuracy is :",acc)
cm=confusion_matrix(ytest,ypred)
sns.heatmap(cm,annot=True)
print(classification_report(ytest,ypred))
```

### Over Sampling 


```python
nonfraud=df[df.Class==0].index
fraud=np.array(df[df.Class==1].index)
random_index=np.array(np.random.choice(fraud,len(nonfraud),replace=True))
oversample_index=np.concatenate([nonfraud,random_index])
oversample=df.iloc[oversample_index,:]
oversample["Class"].value_counts()
```

```python
x=oversample.drop(columns=["Class"])
y=oversample["Class"]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
acc=accuracy_score(ytest,ypred)
print("Accuracy is :",acc)
cm=confusion_matrix(ytest,ypred)
sns.heatmap(cm,annot=True)
print(classification_report(ytest,ypred))
```
